```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
# get data and read
data = 'NUTS_RG_10M_2013_4326_LEVL_1.shp'
pth = os.path.abspath(os.curdir) + os.sep + 'data' + os.sep + data
nuts = gpd.read_file(pth)
```

```python
nuts = nuts.cx[:,30:90]
nuts.plot(figsize=(10, 6))
```

```python
# set index
nuts.set_index('NUTS_ID');
```

```python
nuts.head(3)
```

```python
# plot uk based on string wildcard
uk = nuts[nuts['NUTS_ID'].str.contains("UK")]
uk.plot()
```

```python
# add some random numbers to dataframe
uk['randNumCol'] = np.random.randint(1, 6, uk.shape[0])
uk.plot(column='randNumCol');
```
